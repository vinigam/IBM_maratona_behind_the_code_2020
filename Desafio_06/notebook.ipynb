## Pre-processando o dataset antes do treinamento

### Removendo todas as linhas que possuem algum valor nulos em determinadas colunas

Usando o mtodo Pandas **DataFrame.dropna()** voc pode remover todas as linhas nulas do dataset.

Docs: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html

```python
# Exibindo os dados ausentes do conjunto de dados antes da primeira transformao (df)
print("Valores nulos no df_training_dataset antes da transformao DropNA: \n\n{}\n".format(df_training_dataset.isnull().sum(axis = 0)))
```

```python
# Aplicando a funo para deletar todas as linhas com valor NaN na coluna ``certificados'' e ``total_modulos'':
df_training_dataset = df_training_dataset.dropna(axis='index', how='any', subset=['certificados', 'total_modulos'])
```

```python
# Exibindo os dados ausentes do conjunto de dados aps a primeira transformao (df)
print("Valores nulos no df_training_dataset aps a transformao DropNA: \n\n{}\n".format(df_training_dataset.isnull().sum(axis = 0)))
```

### Processando valores NaN com o SimpleImputer do sklearn

Para os valores NaN, usaremos a substituio pela constante 0 como **exemplo**.

Voc pode escolher a estratgia que achar melhor para tratar os valores nulos :)

Docs: https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html?highlight=simpleimputer#sklearn.impute.SimpleImputer

```python
from sklearn.impute import SimpleImputer
import numpy as np


impute_zeros = SimpleImputer(
    missing_values=np.nan,
    strategy='most_frequent',
    fill_value=0,
    verbose=0,
    copy=True
)
```

```python
# Exibindo os dados ausentes do conjunto de dados antes da primeira transformao (df)
print("Valores nulos no df_training_dataset antes da transformao SimpleImputer: \n\n{}\n".format(df_training_dataset.isnull().sum(axis = 0)))

# Aplicando a transformao ``SimpleImputer`` no conjunto de dados base
impute_zeros.fit(X=df_training_dataset)

# Reconstruindo um Pandas DataFrame com os resultados
df_training_dataset_imputed = pd.DataFrame.from_records(
    data=impute_zeros.transform(
        X=df_training_dataset
    ),
    columns=df_training_dataset.columns
)

# Exibindo os dados ausentes do conjunto de dados aps a primeira transformao (df)
print("Valores nulos no df_training_dataset aps a transformao SimpleImputer: \n\n{}\n".format(df_training_dataset_imputed.isnull().sum(axis = 0)))
```

### Eliminando colunas indesejadas

Vamos **demonstrar** abaixo como usar o mtodo **DataFrame.drop()**.

### Tratamento de de variveis categricas

Como mencionado antes, os computadores no so bons com variveis "categricas" (ou strings).

Dado uma coluna com varivel categrica, o que podemos realizar  a codificao dessa coluna em mltiplas colunas contendo variveis binrias. Esse processo  chamado de "one-hot-encoding" ou "dummy encoding". Se voc no  familiarizado com esses termos, voc pode pesquisar mais sobre isso na internet :)

```python
# Tratando variveis categricas com o mtodo Pandas ``get_dummies()''
df_training = pd.get_dummies(df_training_dataset_rmcolumns, columns=['profissao'])
df_training.tail()
```

# Ateno!

## Treinando um classificador com base em uma rvore de deciso

### Selecionando FEATURES e definindo a varivel TARGET

```python
df_training.columns
```

```python
features = df_training[
    [
        'pretende_fazer_cursos_lit', 'interesse_mba_lit',
        'importante_ter_certificado', 'horas_semanais_estudo', 'total_modulos',
        'modulos_iniciados', 'modulos_finalizados', 'certificados'
        , 'profissao_Advogado', 'profissao_Analista',
        'profissao_Analista Senior', 'profissao_Assessor',
        'profissao_Coordenador', 'profissao_Diretor', 'profissao_Engenheiro',
        'profissao_Gerente', 'profissao_Outros', 'profissao_SEM EXPERINCIA',
        'profissao_Supervisor', 'profissao_Scio/Dono/Proprietrio'
    ]
]
target = df_training['categoria']  ## NO TROQUE O NOME DA VARIVEL TARGET.
```

### Dividindo nosso conjunto de dados em conjuntos de treinamento e teste

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=133)
```

### Treinando uma rvore de deciso

```python
# Mtodo para creacion de modelos basados en arbol de desicin
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

dtc = RandomForestClassifier(max_leaf_nodes = 1000, random_state = 0).fit(X_train, y_train)
```

### Fazendo previses na amostra de teste

```python
y_pred = dtc.predict(X_test)
print(y_pred)
```

### Analisando a qualidade do modelo atravs da matriz de confuso

```python
import matplotlib.pyplot as plt
import numpy as np
import itertools


def plot_confusion_matrix(cm, target_names, title='Confusion matrix', cmap=None, normalize=True):
    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy
    if cmap is None:
        cmap = plt.get_cmap('Blues')
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=45)
        plt.yticks(tick_marks, target_names)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass))
    plt.show()
```

```python
from sklearn.metrics import confusion_matrix


plot_confusion_matrix(confusion_matrix(y_test, y_pred), ['parfil1', 'perfil2', 'perfil3', 'perfil4', 'perfil5', 'perfil6'])
```
